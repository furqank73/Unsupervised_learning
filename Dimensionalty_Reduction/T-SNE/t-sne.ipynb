<div class="tsne-title" style="color:#4F8A8B; font-weight:bold; font-size:2em;">
    T-distributed Stochastic Neighbor Embedding (t-SNE)
</div>
<div class="tsne-desc" style="color:#222; background:linear-gradient(90deg,#f8fafc 60%,#e0e7ff 100%); border-radius:8px; padding:16px;">
    <span class="tsne-keyword" style="color:#b31d28; font-weight:bold;">t-SNE</span> is a 
    <span class="tsne-highlight" style="background:#ffe066; color:#222; border-radius:4px; padding:2px 6px;">non-linear dimensionality reduction</span> 
    technique used for visualizing high-dimensional data in a lower-dimensional space, mainly in 
    <span style="color:#4F8A8B; font-weight:bold;">2D</span> or 
    <span style="color:#4F8A8B; font-weight:bold;">3D</span>.<br><br>
    Unlike linear methods such as 
    <span class="tsne-keyword" style="color:#b31d28; font-weight:bold;">Principal Component Analysis (PCA)</span>, 
    t-SNE focuses on 
    <span class="tsne-highlight" style="background:#ffe066; color:#222; border-radius:4px; padding:2px 6px;">preserving the local structure and patterns</span> 
    of the data, making it especially useful for exploring 
    <span style="color:#7D5FFF; font-weight:bold;">clusters</span> and 
    <span style="color:#00b894; font-weight:bold;">relationships</span> in complex datasets.
</div>

```python
import pandas as pd

#load the dataset
df = pd.read_csv('Mall_Customers.csv')
```

<div style="background: linear-gradient(90deg,rgb(249, 248, 252) 60%, #e0e7ff 100%); border-radius: 12px; padding: 18px; color: #222; font-size: 1.15em;">
    <span style="color:rgb(142, 49, 56); font-weight: bold;">t-SNE</span> works by looking at the <span style="background: #ffe066; color: #222; border-radius: 4px; padding: 2px 6px;">similarity between data points</span> in the high-dimensional space.<br>
    The similarity is computed as a <span style="color: #7D5FFF; font-weight: bold;">conditional probability</span>.<br>
    It calculates how likely it is that one data point would be near another.
</div>

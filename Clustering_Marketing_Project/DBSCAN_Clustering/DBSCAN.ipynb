<h2 style="color:#4034eb; margin-top:0">üåê Density-Based Clustering</h2>
<p style="color:#4a4a4a; line-height:1.6">
    Groups closely-packed points as clusters and identifies isolated points as noise. 
    <strong>Key differentiator:</strong> Discovers natural cluster shapes without requiring pre-defined cluster counts.
</p>
</div>

## <span style="color:#34ebeb; font-weight:600">üîß Key Parameters</span>

<div style="display:grid; grid-template-columns:1fr 1fr; gap:20px; margin:25px 0">

<div style="background:#f3e5f5; padding:15px; border-radius:8px; border-top:3px solid #ab47bc">
<h3 style="color:#8e24aa; margin-top:0">üìè eps (Œµ)</h3>
<ul style="color:#5d5d5d">
    <li>Maximum distance between neighbor points</li>
    <li>Smaller Œµ = more noise detected</li>
    <li><em>Tip:</em> Use k-distance plot to determine</li>
</ul>
</div>

<div style="background:#e8f5e9; padding:15px; border-radius:8px; border-top:3px solid #66bb6a">
<h3 style="color:#2e7d32; margin-top:0">üßÆ min_samples</h3>
<ul style="color:#5d5d5d">
    <li>Minimum points to form a cluster</li>
    <li>Higher values = more robust clusters</li>
    <li><em>Default:</em> 2√ódata dimensions</li>
</ul>
</div>
</div>

## <span style="color:#348feb; font-weight:600">‚ú® Advantages</span>

<div style="background:#7771f5; padding:15px; border-radius:8px; margin:20px 0">
‚úÖ <strong>Shape Flexibility</strong>: Detects non-spherical clusters<br>
‚úÖ <strong>Cluster-Free</strong>: No need to specify cluster count<br>
‚úÖ <strong>Noise Handling</strong>: Explicit outlier identification<br>
‚úÖ <strong>Hierarchy</strong>: Can find nested density patterns
</div>

</h2>

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import DBSCAN

# Select numerical features for clustering
X = df[['Annual Income (k$)', 'Spending Score (1-100)']]

# Standardize features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
df.head()
```

```python
from sklearn.metrics import silhouette_score
import pandas as pd

# Visualize the effect of different min_samples and eps values on DBSCAN clustering


min_samples_grid = list(range(1, 10))
eps_grid = np.arange(0.1, 1.1, 0.1)

results = []

plt.figure(figsize=(50, 30))
plot_num = 1

for ms in min_samples_grid:
    for eps in eps_grid:
        db = DBSCAN(eps=eps, min_samples=ms)
        cluster_labels = db.fit_predict(X_scaled)
        n_clusters = len(set(cluster_labels)) - (1 if -1 in cluster_labels else 0)
        n_noise = list(cluster_labels).count(-1)
        # Silhouette score only if more than 1 cluster and less than all noise
        if n_clusters > 1 and n_noise < len(cluster_labels):
            sil = silhouette_score(X_scaled, cluster_labels)
        else:
            sil = np.nan
        results.append((ms, eps, n_clusters, n_noise, sil))
        
        plt.subplot(len(min_samples_grid), len(eps_grid), plot_num)
        plt.scatter(X['Annual Income (k$)'], X['Spending Score (1-100)'], c=cluster_labels, cmap='Set1', s=10)
        plt.title(f"min_samples={ms}\neps={eps:.1f}\nclusters={n_clusters}\nnoise={n_noise}")
        plt.xticks([])
        plt.yticks([])
        plot_num += 1

plt.tight_layout()
plt.show()

# Show summary table of silhouette scores
results_df = pd.DataFrame(results, columns=['min_samples', 'eps', 'n_clusters', 'n_noise', 'silhouette'])
display(results_df.pivot(index='min_samples', columns='eps', values='silhouette'))
```

    "So, an epsilon of 0.4 with a minimum sample size of 8 seems optimal. Let's apply it and see the results."
</h2>

```python
from sklearn.neighbors import NearestNeighbors

# Compute k-distance (k = min_samples)
min_samples = 8
neigh = NearestNeighbors(n_neighbors=min_samples)
nbrs = neigh.fit(X_scaled)
distances, indices = nbrs.kneighbors(X_scaled)

# Sort and plot the k-distance graph
plt.figure(figsize=(8, 4))
plt.plot(np.sort(distances[:, min_samples - 1]))
plt.title("K-distance Graph (for Epsilon Estimation)")
plt.xlabel("Points sorted by distance")
plt.ylabel(f"{min_samples}-th Nearest Neighbor Distance")
plt.grid(True)
plt.show()
```

```python
# Run DBSCAN
dbscan = DBSCAN(eps=0.4, min_samples=8)
labels = dbscan.fit_predict(X_scaled)

# Add cluster labels to the dataframe
df['DBSCAN_Cluster'] = labels

df['DBSCAN_Cluster'].value_counts()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(
    x=X['Annual Income (k$)'],
    y=X['Spending Score (1-100)'],
    hue=labels,
    palette='Set1',
    legend='full',
    s=60
)
plt.title('DBSCAN Clusters')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.legend(title='Cluster')
plt.show()
```

```python
from sklearn.metrics import silhouette_score

# Compute silhouette score for the current DBSCAN clustering (if more than 1 cluster and not all noise)
if len(set(labels)) > 1 and np.any(labels != -1):
    sil_score = silhouette_score(X, labels)
    print(f"Silhouette Score: {sil_score:.3f}")
else:
    print("Silhouette score is not defined (only one cluster or all points are noise).")
```

# Now lets see Optics

```python
from sklearn.cluster import OPTICS

# Run OPTICS clustering with optimized parameters
optics = OPTICS(min_samples=8, xi=0.05, min_cluster_size=0.05, metric='euclidean')
optics.fit(X_scaled)
optics_labels = optics.labels_

# Add cluster labels to the dataframe
df['Cluster'] = optics.labels_

# Count points in each cluster
cluster_counts = df['Cluster'].value_counts().sort_index()
print("Cluster distribution:")
print(cluster_counts)
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(
    x=X['Annual Income (k$)'],
    y=X['Spending Score (1-100)'],
    hue=optics_labels,
    palette='Set1',
    legend='full',
    s=60
)
plt.title('OPTICS Clusters')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.legend(title='Cluster')
plt.show()
```

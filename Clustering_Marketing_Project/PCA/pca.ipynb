<div class="pca-guide">
  <h2><span class="icon">üß†</span> What is PCA?</h2>

  <p><strong>Principal Component Analysis (PCA)</strong> is a <strong>dimensionality reduction technique</strong>. It transforms your high-dimensional data (e.g., 3 features like Age, Income, Spending) into fewer dimensions (e.g., 2) <strong>while keeping as much information as possible</strong>.</p>

  <div class="highlight-box">
    <p>Think of it as:</p>
    <blockquote>
      A smart way to <strong>compress data</strong>, <strong>remove noise</strong>, and <strong>visualize patterns</strong>, without losing much of what makes the data valuable.
    </blockquote>
  </div>

  <hr>

  <h2><span class="icon">üì¶</span> When to Use PCA?</h2>

  <p>Use PCA when you want to:</p>

  <ul>
    <li><strong>Reduce the number of features</strong> while keeping the structure of the data</li>
    <li><strong>Visualize high-dimensional data</strong> (like converting 3D ‚Üí 2D)</li>
    <li><strong>Find hidden patterns</strong> or groupings in your data</li>
    <li><strong>Speed up machine learning models</strong> by removing redundant features</li>
  </ul>

  <hr>

  <h2><span class="icon">üîç</span> How PCA Works (Step-by-Step)</h2>

  <p>Let's assume you have 3 features: <strong>Age</strong>, <strong>Income</strong>, and <strong>Spending Score</strong>.</p>

  <h3>1. Standardize the Data</h3>
  <p>Why? Because PCA is affected by scale.</p>
  <pre><code>from sklearn.preprocessing import StandardScaler
X_scaled = StandardScaler().fit_transform(X)</code></pre>

  <h3>2. Compute Covariance Matrix</h3>
  <p>This tells how features vary <strong>with respect to each other</strong>.</p>
  <ul>
    <li>If two features vary together, they're redundant</li>
    <li>PCA finds a new axis that removes this redundancy</li>
  </ul>

  <h3>3. Compute Eigenvectors and Eigenvalues</h3>
  <ul>
    <li><strong>Eigenvectors</strong>: Directions of the new axes (called <strong>principal components</strong>)</li>
    <li><strong>Eigenvalues</strong>: How much variance (information) each component captures</li>
  </ul>

  <h3>4. Select Top Principal Components</h3>
  <p>Choose the components that <strong>capture the most variance</strong>:</p>
  <ul>
    <li>PC1 captures 44% of the variance</li>
    <li>PC2 captures 33%</li>
    <li>Together = 77% ‚Üí good enough for 2D visualization</li>
  </ul>

  <h3>5. Project Data Onto the New Axes</h3>
  <p>The original data is transformed (rotated) onto the new PCA axes.</p>

  <hr>

  <h2><span class="icon">üéØ</span> Real-World Analogy</h2>

  <div class="highlight-box">
    <p>Imagine you're looking at a 3D object (like a cube), but you're only allowed to look at it from <strong>one angle</strong>:</p>
    <ul>
      <li>PCA finds the <strong>best angle</strong> that lets you see the <strong>most shape and structure</strong> in just 2D</li>
      <li>It's like taking a <strong>smart photo</strong> of a complex object to keep the most important details</li>
    </ul>
  </div>

  <hr>

  <h2><span class="icon">üìä</span> Interpreting a PCA Plot</h2>

  <p>In your image:</p>
  <ul>
    <li><strong>PC1 (44.3%)</strong>: The main direction along which your data varies</li>
    <li><strong>PC2 (33.3%)</strong>: The second-most important direction</li>
    <li><strong>Red Arrows</strong>: Show how original features (Age, Income, Spending) align with these new directions</li>
  </ul>

  <p>This helps you see <strong>which customers are similar</strong>, and <strong>what features are driving the differences</strong>.</p>

```python
df.shape
```

```python
pca = PCA(n_components=2)
pca_result = pca.fit_transform(df)
```

```python
pca_result.shape
```

```python
print(pca_result)
```

```python
print(f"PC1 explains {pca.explained_variance_ratio_[0]*100:.1f}% of variance")
print(f"PC2 explains {pca.explained_variance_ratio_[1]*100:.1f}% of variance")
```

```python
plt.figure(figsize=(10, 8))
plt.scatter(pca_result[:, 0], pca_result[:, 1], c='dodgerblue', s=200, alpha=0.7, edgecolor='k')

# Annotate points
for i in range(pca_result.shape[0]):
    x, y = pca_result[i, 0], pca_result[i, 1]
    plt.text(x + 0.1, y + 0.1, f'Customer {i}', fontsize=10)

# Draw arrows showing original feature contributions (now 4 features)
features = ['Age', 'Income', 'Spending']
for i, feature in enumerate(features):
    plt.arrow(0, 0, pca.components_[0, i]*3, pca.components_[1, i]*3, 
              color='r', width=0.02, head_width=0.1)
    plt.text(pca.components_[0, i]*3.5, pca.components_[1, i]*3.5, 
             feature, color='red', fontsize=12)

plt.xlabel(f'PC1 ({pca.explained_variance_ratio_[0]*100:.1f}% Variance)', fontsize=12)
plt.ylabel(f'PC2 ({pca.explained_variance_ratio_[1]*100:.1f}% Variance)', fontsize=12)
plt.title('PCA Transformation: 3D ‚Üí 2D Customer Data', pad=20, fontsize=14)
plt.grid(True, alpha=0.3)
plt.axhline(0, color='grey', linestyle='--')
plt.axvline(0, color='grey', linestyle='--')
plt.show()
```

```python
# Before PCA: 3D scatter plot
fig = plt.figure(figsize=(14, 6))

ax1 = fig.add_subplot(1, 2, 1, projection='3d')
ax1.scatter(df['Age'], df['Annual Income (k$)'], df['Spending Score (1-100)'],
            c='dodgerblue', s=60, alpha=0.7)
ax1.set_xlabel('Age')
ax1.set_ylabel('Annual Income (k$)')
ax1.set_zlabel('Spending Score (1-100)')
ax1.set_title('Before PCA: 3D Feature Space')

# Plot eigenvectors as arrows from the mean of the data
mean_vals = df.mean().values
for i in range(2):  # Only first two eigenvectors (since pca.n_components=2)
    vec = pca.components_[i]
    ax1.quiver(
        mean_vals[0], mean_vals[1], mean_vals[2],  # start at mean
        vec[0], vec[1], vec[2],                    # direction
        length=2, color=['red', 'green'][i], linewidth=3, arrow_length_ratio=0.15,
        label=f'PC{i+1} direction'
    )
ax1.legend()

# After PCA: 2D scatter plot
ax2 = fig.add_subplot(1, 2, 2)
ax2.scatter(pca_result[:, 0], pca_result[:, 1], c='orange', s=60, alpha=0.7)

# Annotate each point with CustomerID (use index as CustomerID)
for idx in range(pca_result.shape[0]):
    ax2.annotate(f'ID {idx}', (pca_result[idx, 0], pca_result[idx, 1]), fontsize=7, color='gray')

ax2.set_xlabel(f'PC1 ({pca.explained_variance_ratio_[0]*100:.1f}% variance)')
ax2.set_ylabel(f'PC2 ({pca.explained_variance_ratio_[1]*100:.1f}% variance)')
ax2.set_title('After PCA: 2D Principal Component Space')

# Show original feature names as annotation
ax2.annotate('Original features:\n' + ', '.join(df.columns), 
             xy=(0.05, 1), xycoords='axes fraction', fontsize=10,
             ha='left', va='top', bbox=dict(boxstyle="round,pad=0.3", fc="white", ec="gray", alpha=0.7))

# Add eigenvalues and eigenvectors as text box
eigvals = pca.explained_variance_
eigvecs = pca.components_
eig_text = "Eigenvalues:\n"
for i, val in enumerate(eigvals):
    eig_text += f"PC{i+1}: {val:.3f}\n"
eig_text += "\nEigenvectors:\n"
for i, vec in enumerate(eigvecs):
    eig_text += f"PC{i+1}: [{', '.join(f'{v:.3f}' for v in vec)}]\n"

ax2.annotate(eig_text, xy=(0.98, 0.02), xycoords='axes fraction', fontsize=8,
             ha='right', va='bottom', bbox=dict(boxstyle="round,pad=0.3", fc="white", ec="gray", alpha=0.7))

# Plot eigenvector directions in 2D PCA space
origin_2d = np.zeros(2)
for i in range(2):
    ax2.arrow(origin_2d[0], origin_2d[1], eigvecs[i, 0]*2, eigvecs[i, 1]*2,
              color=['red', 'green'][i], width=0.02, head_width=0.1, label=f'PC{i+1} dir')
ax2.legend()

plt.tight_layout()
plt.show()
```
